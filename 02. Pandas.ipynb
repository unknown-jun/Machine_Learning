## DataFrame, List, Dictionary, Mutual conversion with ndarray
### Transform ndarray, list, dictionary to DataFrame 

```python
import numpy as np

col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape: ', array1.shape, '\n')

# using with list, create DataFrame
df_list1 = pd.DataFrame(list1, columns=col_name1)
print('DataFrame made by one dimention list: \n', df_list1, '\n')

# using with ndarray, create DataFrame
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('DataFrame made by one dimention ndarray: \n', df_array1)
```

```python
# need to three of columns' names
col_name2 = ['col1', 'col2', 'col3']

# Transfrom to DataFrame after create list & ndarray which have shape of 2 rows and x 3 colunms
list2 = [[1, 2, 3],
         [11, 12, 13]]

array2 = np.array(list2)
print('array2 shape: ', array2.shape, '\n')

df_list2 = pd.DataFrame(list2, columns=col_name2)
print('DataFrame made by two dimentions list: \n', df_list2, '\n')

df_array2 = pd.DataFrame(array2, columns=col_name2)
print('DataFrame made by two dimentions ndarray: \n', df_array2)
```

```python
# Key can map with string columns' name
# Value can map with columns' data which has shape of list (or ndarray) 

dict = {'col1':[1, 11], 'col2':[2, 22], 'col':[3, 33]}
df_dict = pd.DataFrame(dict)
print('DataFrame made by dictionary: \n', df_dict)
```

### Transform DataFrame to ndarray, list, dictionary

```python
# Transform DataFrame to ndarray 
array3 = df_dict.values
print('Type of df_dict.values: ', type(array3), 'df_dict.values shape: ', array3.shape)
print(array3)
```

```python
# Transform DataFrame to list
list3 = df_dict.values.tolist()
print('Type of df_list.values.tolist(): ', type(list3))
print(list3)

# Transform DataFrame to dictionary
dict3 = df_dict.to_dict('list')
print('Type of df_dcit.to_dict(): ', type(dict3))
print(dict3)
```

## Index object

```python
# index object can be slicing and returning value
print(type(indexes.values))
print(indexes.values.shape)
print(indexes[:5].values)
print(indexes.values[:5])
print(indexes[6])
```

```python
# Index object is except from when series object apply calculating
series_fair = titanic_df['Fare']
print('Max value of Fair Series: ', series_fair.max())
print('Sum value of Fair Seires: ', series_fair.sum())
print('sum() Fair Series: ', sum(series_fair))
print('Fare Series + 3: \n', (series_fair + 3).head(3))
```
